```python
# ans

MongoDB is a popular NoSQL database that uses a document-oriented data model. It is designed for high performance, high availability, and easy scalability. Instead of storing data in tables and rows as in traditional relational databases, MongoDB stores data in flexible, JSON-like documents, which can vary in structure. These documents are stored in collections, and each document contains key-value pairs that can represent a variety of data types.


What is MongoDB?
MongoDB is a popular NoSQL database that uses a document-oriented data model. It is designed for high performance, high availability, and easy scalability. Instead of storing data in tables and rows as in traditional relational databases, MongoDB stores data in flexible, JSON-like documents, which can vary in structure. These documents are stored in collections, and each document contains key-value pairs that can represent a variety of data types.

Non-Relational Databases (NoSQL)
Non-relational databases, commonly referred to as NoSQL databases, are designed to store, retrieve, and manage data that is modeled in means other than the tabular relations used in relational databases. Key characteristics of NoSQL databases include:

Schema-less: They allow for flexible data structures, meaning that documents in a collection do not need to have a uniform structure.
Scalability: They are designed to scale out by using distributed computing, making it easier to handle large amounts of data across many servers.
Performance: Optimized for high performance on certain types of workloads, such as read-heavy or write-heavy operations.
Types: There are several types of NoSQL databases, including document stores (e.g., MongoDB), key-value stores (e.g., Redis), wide-column stores (e.g., Cassandra), and graph databases (e.g., Neo4j).
Scenarios Where MongoDB is Preferred Over SQL Databases
MongoDB is preferred over traditional SQL databases in several scenarios, including:

Flexible Schema Design:

When the data model is not well-defined or is likely to evolve over time, MongoDB's flexible schema can accommodate changes without requiring significant database refactoring.
High Write Loads:

Applications with high write operations can benefit from MongoDB’s ability to handle high throughput and its capability to distribute writes across multiple nodes in a cluster.
Big Data and Real-Time Analytics:

For applications that require handling large volumes of unstructured or semi-structured data, MongoDB can efficiently store and process this data. It's also suitable for real-time analytics where speed is crucial.
Geographically Distributed Data:

When data needs to be distributed across multiple geographic locations, MongoDB's sharding and replication features ensure high availability and performance across different regions.
Content Management Systems and Catalogs:

Applications like content management systems (CMS), product catalogs, and user profiles, which require storing diverse types of data, benefit from MongoDB's document model.
Internet of Things (IoT):

IoT applications that collect vast amounts of data from various devices can leverage MongoDB’s scalability and flexibility to store heterogeneous data.
Rapid Prototyping and Agile Development:

In scenarios where development speed is crucial, MongoDB's schema-less nature allows developers to quickly iterate on the data model without worrying about schema migrations.
Event Logging and Data Caching:

For applications that need to log events or cache data for fast retrieval, MongoDB’s efficient storage and retrieval mechanisms are advantageous.
```

```python
#ans 4

from pymongo import MongoClient

# Replace <your_connection_string> with your MongoDB connection string
connection_string = "mongodb+srv://<username>:<password>@<cluster-address>/test?retryWrites=true&w=majority"

# Create a MongoClient to the running mongod instance
client = MongoClient(connection_string)

# Connect to the database and collection
db = client["mydatabase"]
collection = db["mycollection"]

# Insert one record
one_record = {
    "name": "Alice Smith",
    "age": 30,
    "address": {
        "street": "456 Elm St",
        "city": "Los Angeles",
        "state": "CA",
        "zip": "90001"
    },
    "interests": ["music", "art", "hiking"]
}

inserted_id_one = collection.insert_one(one_record).inserted_id
print(f"Inserted one document with ID: {inserted_id_one}")

# Find and print the inserted record
record = collection.find_one({"_id": inserted_id_one})
print("Inserted One Record:")
print(record)

# Insert many records
many_records = [
    {
        "name": "Bob Johnson",
        "age": 35,
        "address": {
            "street": "789 Oak St",
            "city": "San Francisco",
            "state": "CA",
            "zip": "94102"
        },
        "interests": ["photography", "travel"]
    },
    {
        "name": "Carol White",
        "age": 28,
        "address": {
            "street": "321 Pine St",
            "city": "Seattle",
            "state": "WA",
            "zip": "98101"
        },
        "interests": ["reading", "gardening"]
    },
    {
        "name": "David Brown",
        "age": 42,
        "address": {
            "street": "654 Maple St",
            "city": "Chicago",
            "state": "IL",
            "zip": "60601"
        },
        "interests": ["cycling", "gaming"]
    }
]

inserted_ids_many = collection.insert_many(many_records).inserted_ids
print(f"Inserted many documents with IDs: {inserted_ids_many}")

# Find and print all inserted records
records = collection.find({"_id": {"$in": inserted_ids_many}})
print("Inserted Many Records:")
for rec in records:
    print(rec)
```

```python
#ans 5

from pymongo import MongoClient

# Replace <your_connection_string> with your MongoDB connection string
connection_string = "mongodb+srv://<username>:<password>@<cluster-address>/test?retryWrites=true&w=majority"

# Create a MongoClient to the running mongod instance
client = MongoClient(connection_string)

# Connect to the database and collection
db = client["mydatabase"]
collection = db["mycollection"]

# Insert sample records
sample_records = [
    {"name": "Alice Smith", "age": 30, "city": "Los Angeles"},
    {"name": "Bob Johnson", "age": 35, "city": "San Francisco"},
    {"name": "Carol White", "age": 28, "city": "Seattle"},
    {"name": "David Brown", "age": 42, "city": "Chicago"}
]
collection.insert_many(sample_records)

# Find all documents
print("All documents:")
for doc in collection.find():
    print(doc)

# Find documents with age greater than 30
print("\nDocuments with age > 30:")
for doc in collection.find({"age": {"$gt": 30}}):
    print(doc)

# Find documents with a specific projection
print("\nDocuments with projection (only name and age):")
for doc in collection.find({"age": {"$gt": 25}}, {"name": 1, "age": 1, "_id": 0}):
    print(doc)

# Find documents sorted by age in ascending order and limit to 2 results
print("\nDocuments sorted by age (ascending) and limited to 2 results:")
for doc in collection.find().sort("age", 1).limit(2):
    print(doc)
```

```python
#ans 6

The sort() method in MongoDB is used to sort the results of a query in either ascending or descending order. This method takes a field or multiple fields by which to sort the documents, and it specifies the sort order for each field.

from pymongo import MongoClient

# Replace <your_connection_string> with your MongoDB connection string
connection_string = "mongodb+srv://<username>:<password>@<cluster-address>/test?retryWrites=true&w=majority"

# Create a MongoClient to the running mongod instance
client = MongoClient(connection_string)

# Connect to the database and collection
db = client["mydatabase"]
collection = db["mycollection"]

# Insert sample records
sample_records = [
    {"name": "Alice Smith", "age": 30, "city": "Los Angeles"},
    {"name": "Bob Johnson", "age": 35, "city": "San Francisco"},
    {"name": "Carol White", "age": 28, "city": "Seattle"},
    {"name": "David Brown", "age": 42, "city": "Chicago"}
]
collection.insert_many(sample_records)

# Sort by age in ascending order
print("Documents sorted by age (ascending):")
for doc in collection.find().sort("age", 1):
    print(doc)

# Sort by age in descending order
print("\nDocuments sorted by age (descending):")
for doc in collection.find().sort("age", -1):
    print(doc)

# Sort by city in ascending order and then by age in descending order
print("\nDocuments sorted by city (ascending) and age (descending):")
for doc in collection.find().sort([("city", 1), ("age", -1)]):
    print(doc)
```
