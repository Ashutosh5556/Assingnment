```python
#ans 2
One-Tailed Test:

Definition: A test where the alternative hypothesis specifies a direction of the effect (e.g., greater than or less than).
Example: Testing whether a new drug increases patient recovery time. Here, the alternative hypothesis might be that the recovery time with the new drug is less than with the current drug.
Application: Use when you are interested in testing for an effect in only one direction.
Two-Tailed Test:

Definition: A test where the alternative hypothesis does not specify a direction, only that there is a difference.
Example: Testing whether a new drug has a different effect on recovery time compared to the current drug, without specifying whether it's faster or slower.
Application: Use when you are interested in detecting an effect in either direction (greater or lesser).
```

```python
#ans 3
Type 1 Error (False Positive):

Definition: Occurs when the null hypothesis (H₀) is true, but you incorrectly reject it.
Example Scenario: A medical test incorrectly indicates that a patient has a disease (when they actually do not). The consequence is potentially starting unnecessary treatment.
Type 2 Error (False Negative):

Definition: Occurs when the null hypothesis (H₀) is false, but you fail to reject it.
Example Scenario: A medical test fails to detect a disease in a patient who actually has it. The consequence is that the patient does not receive the necessary treatment.
```
