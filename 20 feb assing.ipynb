```python
#ans 1

GET Method
Purpose: Requests data from a server.
Data Sent: In the URL (query string).
Visibility: Visible in the URL.
Idempotent: Safe to repeat without changing the server state.
Caching: Can be cached by browsers.
Size Limit: Limited by URL length (typically around 2048 characters).
POST Method
Purpose: Submits data to a server to create or update resources.
Data Sent: In the request body.
Visibility: Not visible in the URL.
Non-idempotent: Repeating can change the server state.
Caching: Not cached by browsers.
Size Limit: No practical limit on data size.
```

```python
#ans 4
In Flask, templates are files (typically HTML) used to dynamically generate web pages. They allow the separation of HTML content from Python code, making the application more organized and easier to maintain. Flask uses the Jinja2 template engine to facilitate this.

The render_template() function is used to render these templates by combining them with data passed from the Flask application. This function selects the appropriate template file, processes it with the provided data, and returns the final HTML content to be sent as the HTTP response to the client.
```

from flask import Flask, request, jsonify

app = Flask(__name__)

@app.route('/api/greet', methods=['GET'])
def greet():
    name = request.args.get('name', 'World')
    return jsonify(message=f'Hello, {name}!')

if __name__ == '__main__':
    app.run(debug=True)
