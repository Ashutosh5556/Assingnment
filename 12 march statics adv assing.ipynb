```python
#ans 2
Given:

Expected distribution: 20% blue, 20% orange, 20% green, 10% yellow, 10% red, 20% brown
Observed counts (you'll need to provide the actual counts from your sample).
import scipy.stats as stats

# Observed counts (example)
observed = [35, 30, 25, 10, 5, 20]

# Expected counts based on percentages
expected = [20, 20, 20, 10, 10, 20]

# Perform chi-square goodness of fit test
chi2_stat, p_value = stats.chisquare(observed, f_exp=expected)

print(f"Chi-Square Statistic: {chi2_stat}")
print(f"P-value: {p_value}")
Interpretation:
The Chi-Square statistic gives the measure of the discrepancy between the observed and expected frequencies.
The p-value helps determine whether to reject the null hypothesis. If the p-value is less than 0.05, reject the null hypothesis, indicating that the distribution of M&Ms differs significantly from the expected distribution.
```

```python
#ans 3
import scipy.stats as stats
import numpy as np

# Contingency table
observed = np.array([[20, 10, 15],
                     [15, 25, 20]])

# Perform the Chi-Square test
chi2_stat, p_value, dof, expected = stats.chi2_contingency(observed)

print(f"Chi-Square Statistic: {chi2_stat}")
print(f"P-value: {p_value}")
print(f"Degrees of Freedom: {dof}")
print(f"Expected Frequencies: \n{expected}")
```

```python
#ans 6
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Degrees of freedom
df = 10

# Generate the chi-square distribution data
x = np.linspace(0, 30, 1000)
y = stats.chi2.pdf(x, df)

# Plot the chi-square distribution
plt.plot(x, y, label=f'Chi-Square Distribution df={df}')

# Shade the area for chi-square statistic = 15
x_shade = np.linspace(15, 30, 1000)
y_shade = stats.chi2.pdf(x_shade, df)
plt.fill_between(x_shade, y_shade, color='red', alpha=0.5, label='Area for Chi-Square Statistic = 15')

# Label the plot
plt.title('Chi-Square Distribution with 10 Degrees of Freedom')
plt.xlabel('Chi-Square Statistic')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#ans 11
import scipy.stats as stats

# Given data
sample_mean = 72
sample_std_dev = 10
sample_size = 30
hypothesized_mean = 70

# Calculate the t-statistic
t_stat = (sample_mean - hypothesized_mean) / (sample_std_dev / (sample_size ** 0.5))

# Calculate the p-value
p_value = 2 * stats.t.sf(abs(t_stat), df=sample_size - 1)

print(f"T-Statistic: {t_stat}")
print(f"P-value: {p_value}")
```
