```python
#ans 2
A scatter plot is a type of data visualization that displays values for typically two variables for a set of data. The data is displayed as a collection of points, each having the value of one variable determining the position on the horizontal axis and the value of the other variable determining the position on the vertical axis. Scatter plots are useful for observing the relationship between two variables and identifying patterns, trends, or correlations.
import numpy as np
import matplotlib.pyplot as plt

# Generate data
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

# Create scatter plot
plt.scatter(x, y)

# Add title and labels
plt.title('Scatter Plot of Generated Data')
plt.xlabel('X values')
plt.ylabel('Y values')

# Show plot
plt.show()
```

```python
#ans 3
#The subplot() function in Matplotlib is used to create multiple plots within a single figure. It allows you to arrange multiple subplots in a grid, making it easier to compare different plots side by side or in a specific layout. This function is useful when you want to visualize multiple datasets simultaneously or present different views of the same data.
import numpy as np
import matplotlib.pyplot as plt

# Data for the plots
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create a figure with 4 subplots
plt.figure(figsize=(10, 8))

# Plot 1
plt.subplot(2, 2, 1)
plt.plot(x1, y1, label='Line 1')
plt.title('Line Plot 1')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.legend()

# Plot 2
plt.subplot(2, 2, 2)
plt.plot(x2, y2, label='Line 2')
plt.title('Line Plot 2')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.legend()

# Plot 3
plt.subplot(2, 2, 3)
plt.plot(x3, y3, label='Line 3')
plt.title('Line Plot 3')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.legend()

# Plot 4
plt.subplot(2, 2, 4)
plt.plot(x4, y4, label='Line 4')
plt.title('Line Plot 4')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.legend()

# Adjust layout
plt.tight_layout()

# Show plot
plt.show()
```

```python
#ans 4
A bar plot, or bar chart, is a graphical representation of data where individual categories are represented by rectangular bars with lengths proportional to the values they represent. Bar plots are commonly used to compare different categories of data, such as frequencies, counts, or other numerical measures.

Why is it Used?
Comparison: Allows for easy comparison between different categories.
Clarity: Provides a clear visual representation of data.
Categorical Data: Ideal for displaying and comparing discrete data points.
import numpy as np
import matplotlib.pyplot as plt

# Data
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Vertical Bar Plot
plt.figure(figsize=(10, 5))
plt.bar(company, profit, color='blue')
plt.title('Company Profits - Vertical Bar Plot')
plt.xlabel('Company')
plt.ylabel('Profit (in millions)')
plt.show()

# Horizontal Bar Plot
plt.figure(figsize=(10, 5))
plt.barh(company, profit, color='green')
plt.title('Company Profits - Horizontal Bar Plot')
plt.xlabel('Profit (in millions)')
plt.ylabel('Company')
plt.show()
```

A box plot, also known as a whisker plot, is a standardized way of displaying the distribution of data based on a five-number summary: minimum, first quartile (Q1), median, third quartile (Q3), and maximum. Box plots can also indicate which observations, if any, might be considered outliers.
Summarize Data: Provides a summary of a datasets distribution at a glance.
Compare Distributions: Useful for comparing distributions across different datasets.
Identify Outliers: Helps identify outliers and variability outside the upper and lower quartiles.
Visualize Spread and Skewness: Shows the spread and skewness in the data.
import numpy as np
import matplotlib.pyplot as plt

# Generate data
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Create box plot
plt.figure(figsize=(10, 6))
plt.boxplot([box1, box2], labels=['Box 1', 'Box 2'])
plt.title('Box Plot of Generated Data')
plt.xlabel('Data Sets')
plt.ylabel('Values')
plt.show()
