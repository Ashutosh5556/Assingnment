```python
#ans 2
import seaborn as sns
import matplotlib.pyplot as plt

# Load the fmri dataset
fmri = sns.load_dataset('fmri')

# Create a line plot
sns.lineplot(data=fmri, x='timepoint', y='signal', hue='event', style='region')

# Display the plot
plt.title('FMRI Signal Over Time by Event and Region')
plt.show()
```

```python
#ans 3
import seaborn as sns
import matplotlib.pyplot as plt

# Load the titanic dataset
titanic = sns.load_dataset('titanic')

# Create a figure with two subplots
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Plot the box plot for 'age' vs 'pclass'
sns.boxplot(data=titanic, x='pclass', y='age', ax=axes[0])
axes[0].set_title('Age Distribution by Passenger Class')
axes[0].set_xlabel('Passenger Class')
axes[0].set_ylabel('Age')

# Plot the box plot for 'fare' vs 'pclass'
sns.boxplot(data=titanic, x='pclass', y='fare', ax=axes[1])
axes[1].set_title('Fare Distribution by Passenger Class')
axes[1].set_xlabel('Passenger Class')
axes[1].set_ylabel('Fare')

# Display the plots
plt.tight_layout()
plt.show()

```

```python
#ans 4
import seaborn as sns
import matplotlib.pyplot as plt

# Load the diamonds dataset
diamonds = sns.load_dataset('diamonds')

# Create a histogram for the 'price' column with 'cut' as the hue
sns.histplot(data=diamonds, x='price', hue='cut', kde=True, element="step", palette="muted")

# Set the title and labels
plt.title('Price Distribution by Cut')
plt.xlabel('Price')
plt.ylabel('Frequency')

# Display the plot
plt.show()
```
